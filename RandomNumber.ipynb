$$ X_{n+1} = (a X_n + b) \mod m $$

````python
def random_generator(a, b, m):
    random_generator.seeds = (a * random_generator.seeds + b ) % m
    return random_generator.seeds


random_generator.seeds = 10001
````

```c
// random.cpp

// Park and Miller's psuedo-random number generator.

#define RANDMAX 2147483646 // RANDMAX = M - 1

int random (int seed = 0)
{
  static int next = 1;
  static int A = 16807;
  static int M = 2147483647;   // 2^31 - 1
  static int q = 127773;       // M / A
  static int r = 2836;         // M % A

  if (seed) next = seed;
  next = A * (next % q) - r * (next / q);
  if (next < 0) next += M;
  return next;
}
```

````python
def pm_psuedo_rand(seed = 0):
    if seed:
        pm_psuedo_rand.next = seed
    
    pm_psuedo_rand.next = pm_psuedo_rand.A * (pm_psuedo_rand.next % pm_psuedo_rand.q) - \
      pm_psuedo_rand.r * (pm_psuedo_rand.next / pm_psuedo_rand.q)
    
    if (pm_psuedo_rand.next < 0 ):
        pm_psuedo_rand.next = pm_psuedo_rand.next + pm_psuedo_rand.M
    
    return pm_psuedo_rand.next


pm_psuedo_rand.next = 1;
pm_psuedo_rand.A = 16807;
pm_psuedo_rand.M = 2147483647;   ## 2^31 - 1
pm_psuedo_rand.q = 127773;       ## M / A
pm_psuedo_rand.r = 2836;         ## M % A
    
````
